````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
# carrega o dataset

df = pd.read_csv('demo01.csv', encoding='utf-8', na_values=('na','NaN','NAN'))

df.head()
````

### A base de dados está completa.

````python
# checar se há dados duplicados

df1 = df.drop_duplicates()

print(f'Pós-check: {len(df1)}, Pré-check: {len(df)}\n')

print(f'{round((len(df)-len(df1))/len(df),2)*100}% dos dados estavam duplicados')
````

````python
df1.info()
````

### Verificando a distribuição da variável resposta:

````python
df_mau = df1[df['mau']==True]
df_bom = df1[df1['mau']==False]
````

### Ajustando o valor das variáveis 'tempo_emprego' e 'idade'

````python
def round_number(valor) -> None:
    
    if valor > 0:
        return round(valor,1)
````

````python
df1['tempo_emprego'] = df1['tempo_emprego'].apply(round_number)
df1['idade'] = df1['idade'].apply(round_number)

df1.head()
````

### Converte variáveis categóricas para 'dummies'

Sexo, 0=Feminino, 1=Masculino

Veículo, 0=Não possui, 1=Possui

Imóvel, 0=Não possui, 1=Possui

````python
# converte as variáveis citadas acima para dummies

df2 = pd.get_dummies(data=df1, columns=['sexo','posse_de_veiculo', 'posse_de_imovel'], drop_first=True)

df2.head(2)
````

````python
# renomeia as colunas que foram alteradas na criação dos dummies

df3 = df2.rename(columns={"sexo_M": "sexo", "posse_de_veiculo_Y": "posse_de_veiculo", "posse_de_imovel_Y": "posse_de_imovel"})
df3.head(2)
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
# cria um df apropriado para o scikitlearn

df4 = df3.select_dtypes(include=(np.number, 'bool'))

df4.head(2)
````
