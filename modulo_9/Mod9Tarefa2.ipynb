```python
sinasc_raw = pd.read_csv('sinasc_RO_2019.csv')
sinasc_raw.head()
```

```python
sinasc_raw.columns
```

```python
def faixas(col:str, valor:float) -> None:

    condicao = sinasc_raw[col] <= valor

    df = sinasc_raw.groupby(by=condicao).mean()
    
    print('media')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])


    df = sinasc_raw.groupby(by=condicao).sum()

    print('soma')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])


    df = sinasc_raw.groupby(by=condicao).min()

    print('minimo')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])


    df = sinasc_raw.groupby(by=condicao).max()

    print('maximo')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])


    df = sinasc_raw.groupby(by=condicao).median()

    print('mediana')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])


    df = sinasc_raw.groupby(by=condicao).std()

    print('std')
    print(df['CONSULTAS'])
    print('\n')
    print(df['PARIDADE'])
    
    return None
```
